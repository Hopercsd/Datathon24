```python
# Import necessary libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
random_seed = random.randint(0, 10000)
print(random_seed)
# Ensure the target column has no missing values
housing = housing.dropna(subset=['Listing.Price.ClosePrice'])

# Select all numeric columns as predictors, excluding the target
numeric_columns = housing.select_dtypes(include=['number']).columns
features = numeric_columns.drop('Listing.Price.ClosePrice')  # Use all numeric columns except the target
target = 'Listing.Price.ClosePrice'

# Drop rows with missing values in selected features
housing = housing.dropna(subset=features)

# Split the data into training and testing sets
train_data = housing.sample(frac=0.8, random_state=1)
test_data = housing.drop(train_data.index)

# Initialize the Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=random_seed)

# Fit the model to the training data
model.fit(train_data[features], train_data[target])

# Make predictions on the test data
predictions = model.predict(test_data[features])

# Evaluate the model's performance
mse = mean_squared_error(test_data[target], predictions)
r2 = r2_score(test_data[target], predictions)

print("Mean Squared Error:", mse)
print("R² Score:", r2)

mae = mean_absolute_error(test_data[target], predictions)
print("Mean Absolute Error (MAE):", mae)

def mean_absolute_percentage_error(y_true, y_pred):
    """
    Calculate MAPE given true and predicted values.
    """
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = mean_absolute_percentage_error(test_data[target], predictions)
print("Mean Absolute Percentage Error (MAPE):", mape, "%")
```

```python
# Import necessary libraries
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Ensure the target column has no missing values
housing = housing.dropna(subset=['Listing.Price.ClosePrice'])

# Select all numeric columns as predictors, excluding the target
numeric_columns = housing.select_dtypes(include=['number']).columns
features = numeric_columns.drop('Listing.Price.ClosePrice')  # Use all numeric columns except the target
target = 'Listing.Price.ClosePrice'

# Drop rows with missing values in selected features
housing = housing.dropna(subset=features)

# Split the data into training and testing sets
train_data = housing.sample(frac=0.8, random_state=1)
test_data = housing.drop(train_data.index)

# Normalize the feature data for neural networks
scaler = StandardScaler()
X_train = scaler.fit_transform(train_data[features])
X_test = scaler.transform(test_data[features])
y_train = train_data[target]
y_test = test_data[target]

# Train a neural network (Multi-Layer Perceptron) model
model = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=1)
model.fit(X_train, y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Evaluate the model's performance
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Mean Absolute Error (MAE):", mae)
print("R² Score:", r2)
mae = mean_absolute_error(test_data[target], predictions)
print("Mean Absolute Error (MAE):", mae)
mape = mean_absolute_percentage_error(test_data[target], predictions)
print("Mean Absolute Percentage Error (MAPE):", mape, "%")
```

```python
# Import necessary libraries
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

random_seed = random.randint(0, 10000)
print(random_seed)
# Initialize the Extra Trees Regressor
et_model = ExtraTreesRegressor(n_estimators=100, random_state = random_seed)

# Train the model on the training data
et_model.fit(train_data[features], train_data[target])

# Make predictions on the test data
et_predictions = et_model.predict(test_data[features])

# Evaluate the model's performance
et_mse = mean_squared_error(test_data[target], et_predictions)
et_r2 = r2_score(test_data[target], et_predictions)
et_mae = mean_absolute_error(test_data[target], et_predictions)

# Define MAPE function
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

et_mape = mean_absolute_percentage_error(test_data[target], et_predictions)

# Display the performance metrics
print("Extra Trees Regressor Performance:")
print("Mean Squared Error (MSE):", et_mse)
print("R² Score:", et_r2)
print("Mean Absolute Error (MAE):", et_mae)
print("Mean Absolute Percentage Error (MAPE):", et_mape, "%")
```

```python
# Import necessary libraries
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define the parameter grid for Extra Trees
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30],
}

# Initialize the Extra Trees Regressor
et = ExtraTreesRegressor(random_state=random.randint(0, 10000))

# Initialize GridSearchCV for fine-tuning
grid_search_et = GridSearchCV(estimator=et, param_grid=param_grid, cv=5, scoring='r2', verbose=2, n_jobs=-1)

# Fit the model to the training data
grid_search_et.fit(train_data[features], train_data[target])

# Display the best parameters
print("Best Parameters for Extra Trees Regressor:", grid_search_et.best_params_)

# Use the best model to make predictions
best_et_model = grid_search_et.best_estimator_
et_predictions = best_et_model.predict(test_data[features])

# Evaluate the Extra Trees model's performance
et_mse = mean_squared_error(test_data[target], et_predictions)
et_r2 = r2_score(test_data[target], et_predictions)
et_mae = mean_absolute_error(test_data[target], et_predictions)

# Define MAPE function
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

et_mape = mean_absolute_percentage_error(test_data[target], et_predictions)

print("\nExtra Trees Regressor Performance After Fine-Tuning:")
print("Mean Squared Error (MSE):", et_mse)
print("R² Score:", et_r2)
print("Mean Absolute Error (MAE):", et_mae)
print("Mean Absolute Percentage Error (MAPE):", et_mape, "%")
```
